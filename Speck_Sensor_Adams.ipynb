Are there standards for outdoor pm2.5 concentrations?
Yes, the United States Environmental Protection Agency (EPA) established National Ambient Air Quality Standards for PM2.5  in 1997 and revised them in 2006. National Ambient Air Standards are  established to be protective of public health. The short-term standard  (24-hour or daily average) is 35 micrograms per cubic meter of air (g/m3) and the long-term standard (annual average) is 15 g/m3. A microgram is a unit of weight. There are a million micrograms in a gram, and a pound is equal to about 450 grams.
